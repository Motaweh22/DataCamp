```python
import matplotlib.pyplot as plt
```

# line plot

```python
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.plot(year,pop)#horizontal is left , verical is right
plt.show
```

# Scatter plot

```python
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.scatter(year,pop)#horizontal is left , verical is right
plt.show
```

# check corelation for x-axis

```python
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.scatter(year,pop)#horizontal is left , verical is right
plt.xscale('log')
plt.show
```

# histogram
### a graph used to represent the frequency distribution of a few data points of one variable

```python
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.hist(pop,3)
plt.show()
plt.clf() #cleans it up
```

# Customization

```python
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.scatter(year,pop)
plt.xlabel("years")
plt.ylabel("population")
plt.title("World population")
plt.show()
plt.clf() #cleans it up
```

# Yticks

```python
year=[1950,1960,1970,1980,1990,2000,2005,2010]
pop=[2.519,3.692,5.263,6.972,1.5,9.5,7,4.5]
plt.scatter(year,pop)
plt.xlabel("years")
plt.ylabel("population")
plt.title("World population")
plt.yticks([0,2,4,6,8,10],
           ['0','2B','4B','6B','8B','10B'])
plt.show()
plt.clf() #cleans it up
```

```python
year=[1950,1960,1970,1980,1990,2000,2005,2010]
pop=[2.519,3.692,5.263,6.972,1.5,9.5,7,4.5]
plt.scatter(year,pop)
plt.xlabel("years")
plt.ylabel("population")
plt.title("World population")
plt.yticks([0,2,4,6,8,10],
           ['0','2B','4B','6B','8B','10B'])

plt.text(1990, 1.5, 'lowest')
plt.text(2000, 9.5, 'Highest')

# Add grid() call
plt.grid(True)
plt.show
```
